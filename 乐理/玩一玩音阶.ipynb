来瞧瞧所有音的所有音程得到的音！

```python
from ykimusic import Pitch, Interval
from prettytable import PrettyTable
 
*roots, = map(Pitch, ['C', 'C#', 'Db', 'D', 'D#', 'Eb', 'E', 'F', 'F#', 'Gb', 'G', 'G#', 'Ab', 'A', 'A#', 'Bb', 'B'])

intervals = 'm2', 'M2', 'm3', 'M3', 'D4', 'P4', 'A4', 'D5', 'P5', 'A5', 'm6', 'M6', 'm7', 'M7', 'D8', 'A8'
tb = PrettyTable(['root', *intervals])
for root in roots:
    pitchs: list = [root]  
    for interval in intervals:
        res = root + interval
        pitchs.append(res if len(str(res.accidental)) <= 1 else '-')
    tb.add_row(pitchs)
tb
```
